## Perform GridSearchCV of the following three models on the provided training set (X_train.csv and y_train.csv)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier

# อ่านข้อมูลมาเก็บไว้ในตัวแปร X_train, y_train, X_test, y_test
X_train = pd.read_csv('X_train.csv', header=None)
y_train = pd.read_csv('y_train.csv', header=None)
X_test = pd.read_csv('X_test.csv', header=None)
y_test = pd.read_csv('y_test.csv', header=None)
```

### Model: AdaBoostClassifier

```python
# เลือกค่า Parameter ของ AdaBoostClassifier ที่ต้องการเพื่อทำ GridSearchCV
param_grid = {
  'n_estimators': [50, 100, 200],
  'learning_rate': [0.01, 0.1, 1.0]
}

model = AdaBoostClassifier() # โมเดล AdaBoostClassifier

grid_search_Ada = GridSearchCV(estimator=model, param_grid=param_grid, cv=5)
grid_search_Ada.fit(X_train, y_train) # Fit GridSearchCV กับข้อมูล Train
```

```python
print("Best AdaBoostClassifier parameters:", grid_search_Ada.best_params_)
print("Best AdaBoostClassifier score:", grid_search_Ada.best_score_)
```

#### 1. Evaluate the AdaBoost model on the test set (X_test.csv and y_test.csv)

```python
best_ada_model = grid_search_Ada.best_estimator_

test_score = best_ada_model.score(X_test, y_test)

print("AdaBoostClassifier test score:", test_score)
```

#### 2. For AdaBoost model, plot the feature importances

```python
from xgboost import XGBClassifier, plot_importance
import matplotlib.pyplot as plt

# เลือก feature ที่สำคัญจากการทำ GridSearchCV
importances = best_ada_model.feature_importances_

# ทำการ Plot feature importances
plt.figure(figsize=(10, 5))
plt.bar(range(X_train.shape[1]), importances)
plt.xticks(range(X_train.shape[1]))
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.title("Feature Importances of AdaBoostClassifier")
plt.show()
```

### Model: XGBClassifier

```python
# เลือกค่า Parameter ของต้นไม้ที่ต้องการในการทำ GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.001]
}

xgb_model = XGBClassifier() # โมเดล XGBClassifier

grid_search_xgb = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5)
grid_search_xgb.fit(X_train, y_train) # Fit GridSearchCV กับ X_train และ y_train
```

```python
print("Best parameters:", grid_search_xgb.best_params_)
print("Best score:", grid_search_xgb.best_score_)
```

#### 1. Evaluate the XGBoost model on the test set (X_test.csv and y_test.csv)

```python
best_xgb_model = grid_search_xgb.best_estimator_

test_score = best_xgb_model.score(X_test, y_test)

print("XGBClassifier test score:", test_score)
```

#### 2. For XGBoost model, plot the feature importances

```python
plot_importance(best_xgb_model)
plt.title("XGBoost Feature Importances")
plt.show()
```

### Model: LGBMClassifier

```python
from lightgbm import LGBMClassifier

# เลือกค่า Parameter ของต้นไม้ที่ต้องการในการทำ GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.001]
}

lgbm_model = LGBMClassifier() # โมเดล LGBMClassifier

grid_search_lgbm = GridSearchCV(estimator=lgbm_model, param_grid=param_grid, cv=5) # Fit GridSearchCV กับ X_train และ y_train
grid_search_lgbm.fit(X_train, y_train)
```

```python
print("Best parameters:", grid_search_lgbm.best_params_)
print("Best score:", grid_search_lgbm.best_score_)
```

#### 1. Evaluate the LightGBM model on the test set (X_test.csv and y_test.csv)

```python
best_lgbm_model = grid_search_lgbm.best_estimator_

test_score = best_lgbm_model.score(X_test, y_test)

print("LGBMClassifier test score:", test_score)
```

#### 2. For LightGBM model, plot the feature importances

```python
from lightgbm import plot_importance
import matplotlib.pyplot as plt

plot_importance(best_lgbm_model)
plt.title("LightGBM Feature Importances")
plt.show()
```
