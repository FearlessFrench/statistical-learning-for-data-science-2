```python
# กำหนดตัวแปร tfidf
tfidf = TfidfVectorizer(stop_words="english")
Xtrain_tfidf = tfidf.fit_transform(Xtrain)
Xtest_tfidf = tfidf.transform(Xtest)
Xtrain_tfidf
```

```python
# สร้าง pipeline สำหรับทำ Grid Search
pipeline = Pipeline([('tfidf', TfidfVectorizer()),
                     ('nb', MultinomialNB())])
pipeline.fit(Xtrain, ytrain)
ypred = pipeline.predict(Xtest)

print(classification_report(ytest, ypred))
```

```python
# ทำ Grid Search
# parameters = {"ชื่อโมเดล__ชื่อพารามิเตอร์ที่เรากำหนดไว้": [พารามิเตอร์]}
params = {"tfidf__stop_words": ['english'],
          "nb__alpha": [0.1, 1, 10, 50, 100]}
gridcv = GridSearchCV(pipeline, params,
                      scoring='f1_macro', cv=5)
gridcv.fit(Xtrain, ytrain)
```

```python
gridcv.best_estimator_
```

```python
print(gridcv.best_params_)
print(gridcv.best_score_)
```

```python
# Make predictions on the test set
ypred = gridcv.predict(Xtest)

# Report classification scores
print(classification_report(ytest, ypred))
```

```python
# สร้าง pipeline ใหม่เพื่อทำ Random Search
pipeline = Pipeline([('nb', MultinomialNB())])

random_params = {"nb__alpha": uniform(loc=0, scale=100)}
clf = RandomizedSearchCV(pipeline, random_params, n_iter=5,
                         scoring='f1_macro', cv=5, error_score='raise')
clf.fit(Xtrain_tfidf, ytrain)
```

```python
print(clf.best_params_)
print(clf.best_score_)
```

```python
clf.best_estimator_
```

```python
# Make predictions on the test set
ypred = clf.predict(Xtest_tfidf)

# Report classification scores
print(classification_report(ytest, ypred))
```
